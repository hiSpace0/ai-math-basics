```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 定义 x 轴范围
x = np.linspace(-5, 5, 1000)

# 定义三个函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def gauss(x):
    return np.exp(-x**2)

# 计算函数值
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_gauss = gauss(x)

# 创建图形
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# 绘制 Sigmoid
ax1.plot(x, y_sigmoid, 'b-', linewidth=2, label='Sigmoid')
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axhline(y=1, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('σ(x)')
ax1.set_title('Sigmoid 函数')
ax1.legend()
ax1.set_ylim([-0.1, 1.1])

# 绘制 Tanh
ax2.plot(x, y_tanh, 'r-', linewidth=2, label='Tanh')
ax2.axhline(y=-1, color='k', linestyle='-', alpha=0.3)
ax2.axhline(y=1, color='k', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x')
ax2.set_ylabel('tanh(x)')
ax2.set_title('Tanh 函数')
ax2.legend()
ax2.set_ylim([-1.1, 1.1])

# 绘制 Gauss
ax3.plot(x, y_gauss, 'g-', linewidth=2, label='Gauss')
ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x')
ax3.set_ylabel('G(x)')
ax3.set_title('高斯函数 (Gauss)')
ax3.legend()
ax3.set_ylim([-0.1, 1.1])

plt.tight_layout()
plt.savefig('激活函数对比.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 也可以绘制在同一张图上对比
fig2, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y_sigmoid, 'b--', linewidth=2, label='Sigmoid', alpha=0.7)
ax.plot(x, y_tanh, 'r-.', linewidth=2, label='Tanh', alpha=0.7)
ax.plot(x, y_gauss, 'g:', linewidth=2, label='Gauss', alpha=0.7)
ax.grid(True, alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Sigmoid、Tanh 和 Gauss 函数对比')
ax.legend(fontsize=12)
ax.set_xlim([-5, 5])
ax.set_ylim([-1.1, 1.1])

plt.tight_layout()
plt.savefig('三个函数对比.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def generate_pascals_triangle_comb(n):
    """
    使用组合数公式生成杨辉三角
    """
    triangle = []
    
    for i in range(n):
        row = []
        for j in range(i + 1):
            # C(i, j) = i! / (j! * (i-j)!)
            comb = math.comb(i, j)  # Python 3.8+ 内置函数
            # 或者用以下公式：
            # comb = math.factorial(i) // (math.factorial(j) * math.factorial(i - j))
            row.append(comb)
        triangle.append(row)
    
    return triangle

# 生成前8行
triangle = generate_pascals_triangle_comb(8)
for i, row in enumerate(triangle):
    print(f"第{i}行: {row}")
```

```python
n = int(input("请输入 n: "))

sum_factorial = 0
factorial = 1  # 当前阶乘值

for i in range(1, n + 1):
    factorial *= i  # i! = (i-1)! * i
    sum_factorial += factorial
    print(f"{i}! = {factorial}, 当前和 = {sum_factorial}")

print(f"S = 1! + 2! + ... + {i}! = {sum_factorial}")
```
