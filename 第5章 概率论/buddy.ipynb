```python
import random
import numpy as np
import matplotlib.pyplot as plt
from math import exp, factorial

# 让图表能正常显示中文（Windows 用 SimHei）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
n = 50
# 1 表示正面, 0 表示反面
flips = [random.choice([1, 0]) for _ in range(n)]

cum_freq = []          # 每一步的正面累计频率
heads = 0
for i, f in enumerate(flips, 1):
    heads += f
    cum_freq.append(heads / i)

plt.figure(figsize=(8, 5))
plt.plot(range(1, n + 1), cum_freq, marker='o', color='steelblue', label='正面累计频率')
plt.axhline(0.5, color='red', linestyle='--', label='理论概率 0.5')
plt.xlabel('抛硬币次数')
plt.ylabel('正面频率')
plt.title('抛硬币 50 次：正面频率随次数变化')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
lambdas = [1, 2, 5, 7, 10]
max_k = 25   # 横轴取到足够大的 k，确保覆盖各 λ 的尾部

plt.figure(figsize=(10, 6))
for lam in lambdas:
    ks = np.arange(0, max_k + 1)
    # 泊松 PMF: P(X=k) = e^{-λ} λ^k / k!
    probs = np.array([exp(-lam) * lam**k / factorial(k) for k in ks])
    plt.plot(ks, probs, marker='o', linewidth=1.5, label=f'λ={lam}')

plt.xlabel('k（发生次数）')
plt.ylabel('P(X=k)')
plt.title('泊松分布概率质量函数（λ = 1, 2, 5, 7, 10）')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
